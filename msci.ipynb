```python
def softmax(x):
  exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
  return exp_x / np.sum(exp_x, axis=1, keepdims=True)
```

```python
def compute_multinomial_gradient(inputs, weights, labels):
    m, n = inputs.shape
    _, k = weights.shape

    # Compute softmax probabilities
    probabilities = softmax(np.dot(inputs, weights))

    # Compute the gradient
    gradient = np.dot(inputs.T, probabilities - labels) / m

    return gradient

def compute_multinomial_hessian_gradient(inputs, weights, labels, lambda_hyperparam):
    m, n = inputs.shape
    _, k = weights.shape

    probabilities = softmax(np.dot(inputs, weights))
    diag_probabilities = np.diag(probabilities[:, 0])

    hessian = np.dot(np.dot(inputs.T, diag_probabilities), inputs) / m + lambda_hyperparam * np.identity(n)

    # Compute the gradient
    gradient = compute_multinomial_gradient(inputs, weights, labels)

    return hessian, gradient
```

```python
def predict_multinomial_regression(inputs, weights):
  """
  Predict labels for inputs using multinomial logistic regression.

  Parameters:
  - inputs: Input data (shape: m x n, where m is the number of samples and n is the number of features)
  - weights: Weight matrix (shape: n x k, where k is the number of classes)

  Returns:
  - predicted_labels: Predicted labels (shape: m,)
  """
  scores = np.dot(inputs, weights)
  probabilities = softmax(scores)

  # Assign the label with the highest probability for each sample
  predicted_labels = np.argmax(probabilities, axis=1)

  return predicted_labels
```

```python
def eval_multinomial_regression(inputs, weights, labels):
  predicted_probabilities = predict_multinomial_regression(inputs, weights)
  log_probabilities = -np.log(predicted_probabilities[np.arange(len(labels)), labels])
  neg_log_prob = np.mean(log_probabilities)
  accuracy = np.mean(np.argmax(predicted_probabilities, axis=1) == labels)
  return neg_log_prob, accuracy
```

```python
def initialize_weights(n_features, n_classes):
  random_weights = np.random.default_rng().uniform(low=-0.01, high=0.01, size=(n_features, n_classes))
  return random_weights
```

```python
# gradient descent
def train_multinomial_regression_gradient(train_inputs, train_labels, lambda_hyperparam, max_iters=10000, gradient_norm_threshold=0.1, learning_rate=0.0001):
  n_iters = 0
  weights = initialize_weights(train_inputs.shape[1])
  gradient_norm = np.Inf

  while n_iters < max_iters and gradient_norm >= gradient_norm_threshold:
    gradient = compute_multinomial_gradient(train_inputs, weights, train_labels) + lambda_hyperparam * weights
    gradient_norm = np.linalg.norm(gradient)
    weights = weights - learning_rate * gradient

    n_iters += 1

  return weights, n_iters

# newton's method
def train_multinomial_regression_newton(train_inputs, train_labels, lambda_hyperparam, max_iters=10000, gradient_norm_threshold=0.1):
  n_iters = 0
  weights = initialize_weights(train_inputs.shape[1])
  gradient_norm = np.Inf

  while n_iters < max_iters and gradient_norm >= gradient_norm_threshold:
    hessian, gradient = compute_multinomial_hessian_gradient(train_inputs, weights, train_labels, lambda_hyperparam)
    gradient_norm = np.linalg.norm(gradient)
    weights -= np.matmul(np.linalg.inv(hessian), gradient)

    n_iters += 1

  return weights, n_iters
```

```python
def k_folds_split(data, labels, k_folds):
  m = len(data)
  n = len(labels)
  data_fold_size = m // k_folds
  labels_fold_size = n // k_folds

  for i in range(k_folds):
    train_data = np.concatenate([ data[ :i*data_fold_size], data[(i+1)*data_fold_size: ] ])
    train_label = np.concatenate([ labels[ :i*labels_fold_size], labels[(i+1)*labels_fold_size: ] ])
    test_data = data[i*data_fold_size : (i+1)*data_fold_size]
    test_label = labels[i*labels_fold_size : (i+1)*labels_fold_size]
    yield train_data, train_label, test_data, test_label


def cross_validation_multinomial_regression(k_folds, hyperparameters, inputs, labels, algorithm):
  # best_accuracy = 0
  best_hyperparam = 0
  best_neg_log_prob = np.Inf
  neg_log_probabilities = np.array([])

  for hyperparam in hyperparameters:
    curr_neg_log_probs = np.array([])
    curr_accuracies = np.array([])

    for train_data, train_label, test_data, test_label in k_folds_split(inputs, labels, k_folds):
      weights, n_iters = train_multinomial_regression_gradient(train_data, train_label, hyperparam) if algorithm == "gradient descent" else train_multinomial_regression_newton(train_data, train_label, hyperparam)
      neg_log_prob, accuracy = eval_multinomial_regression(test_data, weights, test_label)
      curr_neg_log_probs = np.append(curr_neg_log_probs, neg_log_prob)
      curr_accuracies = np.append(curr_accuracies, accuracy)

    avg_neg_log_prob = np.average(curr_neg_log_probs)
    # avg_accuracy = np.average(curr_accuracies)
    neg_log_probabilities = np.append(neg_log_probabilities, avg_neg_log_prob)

    if avg_neg_log_prob < best_neg_log_prob:
      # best_accuracy = avg_accuracy
      best_hyperparam = hyperparam
      best_neg_log_prob = avg_neg_log_prob

  return best_hyperparam, best_neg_log_prob, neg_log_probabilities
```
